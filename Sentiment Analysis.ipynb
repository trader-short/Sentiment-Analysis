```python
import pandas as pd
import numpy as np
import time 
import datetime as dt
import requests 
import newspaper
from newspaper import Article
from textblob import Word
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer as si
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
```

```python
url_data = pd.read_excel('URLs.xlsx')
```

```python
url_data.head()
```

```python
col = url_data.columns.get_loc('url')
url_data['polarity'] = ''
url_data['subjectivity'] = ''
url_data['vader sentiment'] = ''
for i in range(len(url_data)):
    url = url_data.iloc[i,col]
    article = Article(url)
    article.download()
    article.parse()
    para = article.text
    url_data.iloc[i,url_data.columns.get_loc('polarity')] = TextBlob(para).sentiment[0]
    url_data.iloc[i,url_data.columns.get_loc('subjectivity')] = TextBlob(para).sentiment[1]
    si = SentimentIntensityAnalyzer()
    url_data.iloc[i,url_data.columns.get_loc('vader sentiment')] = si.polarity_scores(para)['compound']   
```
